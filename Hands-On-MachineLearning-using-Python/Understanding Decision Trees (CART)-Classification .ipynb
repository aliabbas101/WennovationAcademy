# CART (Classification and Regression Trees) 
CART is an alternative decision tree building algorithm. It can handle both classification and regression tasks. This algorithm uses a new metric named gini index to create decision points for classification tasks.

# Formulas in CART


## Steps to pick a decision Node
1) Calculate Gini Index for the each attribute <br>
2) Calculate weighted sum of gini indexes for that feature<br>
3) Pick the attribute with lowest gini index value<br>
4) Repeat 1,2,3 until a generalized tree has been created<br>
